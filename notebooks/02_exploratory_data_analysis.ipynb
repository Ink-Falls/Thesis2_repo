```python
# Core Libraries
import pandas as pd
import numpy as np
import os

# Visualization Libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set a professional visual style for all plots
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (12, 7)  # Set default figure size

print("Setup Complete. Libraries loaded.")
```

```python
# Define the path to the dataset
baseline_path = os.path.join("..", "data", "processed", "baseline.csv")

# Load the data
try:
    df = pd.read_csv(baseline_path)
    print(
        f"Successfully loaded {os.path.basename(baseline_path)} with shape: {df.shape}"
    )
except FileNotFoundError:
    print(f"ERROR: File not found at {baseline_path}")

# --- Initial Survey ---
print("\n--- First 5 Rows ---")
display(df.head())

print("\n--- Data Types and Non-Null Counts ---")
df.info()

print("\n--- Statistical Summary of Numerical Features ---")
display(df.describe())
```

```python
# Calculate and print the distribution
print("--- Target Variable 'Result' Distribution ---")
class_counts = df["Result"].value_counts()
class_percentages = df["Result"].value_counts(normalize=True) * 100
print(class_counts)
print(f"\nPositive Cases: {class_percentages['positive']:.2f}%")
print(f"Negative Cases: {class_percentages['negative']:.2f}%")

# Visualize the distribution
plt.figure()
sns.countplot(x="Result", data=df, palette=["#4374B3", "#FF8F33"])
plt.title("Distribution of Dengue Test Results", fontsize=16)
plt.xlabel("Result", fontsize=12)
plt.ylabel("Number of Patients", fontsize=12)
plt.show()
```

```python
# List of key numerical features to analyze
features_to_plot = ["Age", "HCT", "TotalPlateletCountcumm"]

print("--- Distribution of Key Numerical Features ---")

for feature in features_to_plot:
    plt.figure(figsize=(14, 5))

    # Histogram with Kernel Density Estimate
    plt.subplot(1, 2, 1)
    sns.histplot(df[feature], kde=True, bins=30)
    plt.title(f"Distribution of {feature}", fontsize=14)

    # Box Plot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[feature])
    plt.title(f"Box Plot of {feature}", fontsize=14)

    plt.tight_layout()
    plt.show()
```

```python
print("--- Bivariate Analysis: Features vs. Dengue Result ---")

# --- Numerical Features vs. Target ---
for feature in features_to_plot:
    plt.figure()
    sns.boxplot(x="Result", y=feature, data=df, palette=["#4374B3", "#FF8F33"])
    plt.title(f"{feature} by Dengue Result", fontsize=16)
    plt.show()

# --- Categorical Features vs. Target ---
plt.figure()
sns.countplot(x="Gender", hue="Result", data=df, palette=["#4374B3", "#FF8F33"])
plt.title("Dengue Result Distribution by Gender", fontsize=16)
plt.show()
```

```python
# Select only numerical columns for the correlation matrix
numerical_df = df.select_dtypes(include=np.number)

# Calculate the correlation matrix
corr_matrix = numerical_df.corr()

# Plot the heatmap
plt.figure(figsize=(18, 15))
sns.heatmap(
    corr_matrix, cmap="coolwarm", annot=False
)  # annot=True is too cluttered for this many features
plt.title("Correlation Matrix of Numerical Features", fontsize=18)
plt.show()
```
